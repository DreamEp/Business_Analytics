# Cleaning the datas

```
#Delete duplicated rows
clean_data = initial_datas.drop_duplicates()
clean_data
```

```
#review the type of each columns
clean_data['ID'] = clean_data['ID'].astype('category')
clean_data['Reason for absence'] = clean_data['Reason for absence'].astype('category')
clean_data['Month of absence'] = clean_data['Month of absence'].astype('category')
clean_data['Day of the week'] = clean_data['Day of the week'].astype('category')
clean_data['Seasons'] = clean_data['Seasons'].astype('category')
clean_data['Disciplinary failure'] = clean_data['Disciplinary failure'].astype(bool)
clean_data['Education'] = clean_data['Education'].astype('category')
clean_data['Social drinker'] = clean_data['Social drinker'].astype(bool)
clean_data['Social smoker'] = clean_data['Social smoker'].astype(bool)

#rename the column "son"
clean_data.rename({'Son': 'Children'}, axis=1, inplace=True)

#delete each report of absenteism when the value of the absence = 0
clean_data = clean_data[clean_data['Reason for absence'] != 0]

```

```
#Get the informations after cleaning the datas
clean_data.info()
```

```
abs_count = clean_data.groupby("ID").ID.count().sort_values(ascending = False)
abs_count.iloc[:5]
```
